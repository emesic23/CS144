```python
import networkx as nx
import numpy as np
```

# Showing Degree Centrality =/= Betweenness Centrality =/= Closeness Centrality

```python
G = nx.Graph()
for i in range(1, 5, 1):
    G.add_edge(i, i+1)
G.add_edge(6, 2)
G.add_edge(2, 7)
G.add_edge(4, 8)
G.add_edge(4, 9)
G.add_edge(6, 3)
```

```python
nx.draw(G, with_labels=True)
```

```python
sorted(nx.closeness_centrality(G).items(), key=lambda x: x[1])
```

```python
sorted(nx.betweenness_centrality(G).items(), key=lambda x: x[1])
```

```python
sorted(nx.degree_centrality(G).items(), key=lambda x: x[1])
```

Here specifically we see a strong counter-example where nodes 3, 2, 4 dont agree with each other between the 3 centrality measures.
